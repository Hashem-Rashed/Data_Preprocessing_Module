###  To fix this issue You will fill these values with median  for numeric and mode for categorical

```python
# trying to fill missing values in my datset using for loop 
print("Filling missing values...")

# Fill numeric columns with median
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
for col in numeric_cols:
    if df[col].isnull().sum() > 0:
        df[col].fillna(df[col].median(), inplace=True)
        print(f"Filled {col} with median: {df[col].median()}")

# Fill categorical columns with mode
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    if col != 'bike_share_for_all_trip' and df[col].isnull().sum() > 0:
        mode_value = df[col].mode()[0] if not df[col].mode().empty else 'Unknown'
        df[col].fillna(mode_value, inplace=True)
        print(f"Filled {col} with mode: {mode_value}")

print("\nMissing values filled.")
```

```python
#check the Nulls 
print("Missing values after filling:")
print(df.isnull().sum())
print(f"\nTotal missing values remaining: {df.isnull().sum().sum()}")
```

```python
## step 3 ) Good factor for marketing work as in ((`member_birth_year`)) : so need to clean it.
## To know the target audience's age and serve on them

# create a column for users' age depending on the current year


current_year = 2024
df['member_age'] = current_year - df['member_birth_year']

print("Age column created!")
print(f"Range of ages: {df['member_age'].min()} to {df['member_age'].max()}")
```

```python
## check your work 
## check your work
print("Age column statistics:")
print(df['member_age'].describe())
print(f"\nSample ages: {df['member_age'].head(10).tolist()}")
```

```python
## change the new column age data type from float to int

df['member_age'] = df['member_age'].astype(int)
print("Age converted to integer!")
print(f"Age dtype: {df['member_age'].dtype}")
```

```python
## create a function to know the average users ages
def analyze_user_ages():
    avg_age = df['member_age'].mean()
    median_age = df['member_age'].median()
    mode_age = df['member_age'].mode()[0]
    
    print(f"Average age: {avg_age:.1f} years")
    print(f"Median age: {median_age:.1f} years")
    print(f"Most common age: {mode_age:.0f} years")
    
    # Age groups
    bins = [0, 18, 25, 35, 45, 55, 65, 100]
    labels = ['<18', '18-24', '25-34', '35-44', '45-54', '55-64', '65+']
    df['age_group'] = pd.cut(df['member_age'], bins=bins, labels=labels, right=False)
    
    print("\nAge Group Distribution:")
    print(df['age_group'].value_counts().sort_index())
    
    return avg_age, median_age, mode_age

# Call the function
avg_age, median_age, mode_age = analyze_user_ages()
```

# <a id="1"> 1 ) . What is the distribution for User type?</a>

```python
## I will import seaborn library to support me in making visualizations
import seaborn as sns

plt.figure(figsize=(10, 5), dpi=200)
plot = sns.countplot(data=df, x='user_type', palette='Set2')

# Add percentage labels
total = len(df)
for p in plot.patches:
    percentage = f'{100 * p.get_height()/total:.1f}%'
    plot.annotate(percentage, 
                  (p.get_x() + p.get_width()/2., p.get_height()), 
                  ha='center', va='center', 
                  xytext=(0, 10), 
                  textcoords='offset points',
                  fontsize=10)

plt.xlabel('User Type', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.title("Distribution of User Types", fontsize=14, fontweight='bold')
plt.grid(axis='y', alpha=0.3)
plt.show()

# Calculate and print statistics
print("User Type Distribution:")
user_stats = df['user_type'].value_counts()
for user_type, count in user_stats.items():
    percentage = (count / total) * 100
    print(f"{user_type}: {count} users ({percentage:.1f}%)")
```

```python
# trying to fill missing values in my datset using for loop 
for column in df.columns:
    if df[column].isnull().any():
        if df[column].dtype in ['int64','float64']:
            df[column].fillna(df[column].median(), inplace = True)
        else:
            df[column].fillna(df[column].mode()[0], inplace = True)
```

# <a id="4">  4 ) . Do the users' ages have outlier values?</a>


```python
plt.figure(figsize=(5, 2))
sns.boxplot(df['age'],orient='h')
plt.xlabel('Age')
plt.title("Box Plot of Users' Ages ")
plt.show()
```

```python
q1 = df['age'].quantile(0.25)
q3 = df['age'].quantile(0.75)
iqr = q3 - q1

lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

lower_outliers = df[df['age'] < lower_bound].values
upper_outliers = df[df['age'] > upper_bound].values

outliers = df[(df['age'] < lower_bound) | (df['age'] > upper_bound)]

print(f"Number of age outliers: {len(outliers)}")
print(f"Percentage of outliers: {(len(outliers)/len(df)*100):.2f}%")
print(f"Age range of outliers: {outliers['age'].min()} to {outliers['age'].max()}")
```

```python
df = df[(df['age'] >= lower_bound) & (df['age'] <= upper_bound)]
```

```python
plt.figure(figsize=(5, 2))
sns.boxplot(df['age'],orient='h')
plt.xlabel('Age')
plt.title("Box Plot of Users' Ages ")
plt.show()
```

> ### <mark>Your analsys and exploration here </mark>

Yes, there are outliers in users' ages. The box plot shows several outliers above 70 years, with some even exceeding 100 years. These could be data entry errors or legitimate elderly users. About 2.5% of users are considered outliers based on the IQR method.

# <a id="11">  11 ) . How the user distribution depending on their Age and user_type ? </a>

```python
# Make visualization to easy get insight

plt.figure(figsize=(14, 6))

age_bins = [0, 20, 30, 40, 50, 60, 100]
age_labels = ['<20', '20-29', '30-39', '40-49', '50-59', '60+']
df['age_group'] = pd.cut(df['age'], bins=age_bins, labels=age_labels)

plt.subplot(1, 2, 1)
age_user_cross = (pd.crosstab(df['age_group'], df['user_type'], normalize='index') * 100)
age_user_cross = age_user_cross.reindex(age_labels)

colors = sns.color_palette("Set2", n_colors=age_user_cross.shape[1])
age_user_cross.plot(kind='bar',stacked=True,ax=plt.gca(),color=colors)
plt.title("User Type Distribution by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Percentage")
plt.legend(title="User Type")

for i, age in enumerate(age_user_cross.index):
    cumulative = 0
    for user_type in age_user_cross.columns:
        value = age_user_cross.loc[age, user_type]
        plt.text(i, cumulative + value/2, f"{value:.1f}%",
                 ha='center', va='center', color='white', fontsize=9)
        cumulative += value

plt.subplot(1, 2, 2)
sns.boxplot(data=df,x='user_type',y='age',palette='Set2')
plt.title("Age Distribution by User Type")
plt.xlabel("User Type")
plt.ylabel("Age")

plt.tight_layout()
plt.show()
```

> ### <mark>Your analsys and exploration here </mark>

Subscribers are predominantly concentrated in the middle age groups (20â€“49), while customers are more prevalent in younger and older age ranges. Additionally, subscribers show a more consistent age distribution, whereas customers span a wider age range.
